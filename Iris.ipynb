```
independent=dataset.drop('Species',axis='columns')
dependent=dataset.iloc[:,4:]
```

```
from sklearn.linear_model import LogisticRegression
regr=LogisticRegression()
```

```
from sklearn.model_selection import train_test_split
independent_train,independent_test,dependent_train,dependent_test=train_test_split(independent,dependent,test_size=0.33)
```

```
regr.fit(independent_train,dependent_train)
```

```
pred=regr.predict(independent_test)
pred=pd.DataFrame(pred)
pred
```

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dependent_test,pred)
#sns.heatmap(cm,annot=True,linewidths=0.06)
```
